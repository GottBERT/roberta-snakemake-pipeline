```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle("ratios.pkl")
df = df.iloc[:, 3:]
df = df.astype(float)
```

```python
df.describe()
```

```python
for col in df.columns:
    fig, ax = plt.subplots()
    ax.set_title(col)
    df[col].plot(ax=ax, kind='kde')
```

```python
for col in df.columns:
    fig, ax = plt.subplots()
    ax.set_title(col)
    df[col].hist(bins=20, ax=ax)
```

```python
for col in df.columns:
    fig, ax = plt.subplots()
    ax.set_title(col)
    df.boxplot(column=col, ax=ax)
```
